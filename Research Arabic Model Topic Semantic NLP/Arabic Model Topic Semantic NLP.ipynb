# Topic Modeling And Topic Stability

**Common challenge Topic clustering is selecting an appropriate number of topics for a particular data set. Choosing too few topics will produce results that are overly broad, while choosing too many will result in the over-clustering of a data into many redundant, highly-similar topics**

**We want to know how many topics are important to us and after survey to select the appropriate method in such task is measure  coherence score based on singular value decomposition (SVD) clustering.**

**In order to build SVD matrix for our documents we found SVD properties is that the recent studies indicate that SVD is mostly useful for small homogeneous data collections. For large inhomogeneous datasets, the performance of the SVD based text retrieval technique may deteriorate. We propose to partition a large inhomogeneous dataset into several smaller ones with clustered structure. Our experimental results show the clustered SVD strategies may enhance the retrieval accuracy and reduce the computing and storage costs [1].**

**[1]Referance [ Clustered SVD strategies in latent semantic indexing](https://www.sciencedirect.com/science/article/abs/pii/S0306457304001384)**


**To achieve this gaol to apply SVD clustering for small homogeneous data we proposed a solution achieved by Appling standard clustering algorithms in the beginning by generating the vectors by using Word2Vec for a list of sentences**

**We note it is better to use dimension reduction algorithms for vectors length our case is 300 before apply standard clustering algorithm. here we used Principal component analysis (PCA) which optimize the clustering result and that minimizes the cluster-specific variance and that we found is useful before applying K-means clustering algorithm [2][3] so we redact the vector length from dimension 300 into small .by covering 90 percentage in variance of original data  And then we have to identify the number of appropriate clusters used Elbow method**


**[2]Referance [Which would you use first, K-Means Clustering, or Principal Component Analysis](https://www.researchgate.net/post/Which-would-you-use-first-K-Means-Clustering-or-Principal-Component-Analysis)**

**[3]Referance [K-means Clustering via Principal Component Analysis](https://icml.cc/Conferences/2004/proceedings/papers/262.pdf)**

**First Step Calulating the co - occurance matrix and then apply dimmension reduction using Truncated SVD and calculate singular values of svd and fimally select best number of topic based on the strength for each singular value where each value indicate the strength for each topic**

**And the second step to apply measure coherence score we want to identify which cluster is distinguished by small homogeneous  cluster We measure entropy value and select the cluster with  highest entropy value to apply semantic topic modeling based on SVD and apply coherence model to get the score of number of topics**

**And the last step select the smallest number from frist& second steps and use this number for further processing**

```python
def remove_longation(text):
    # remove tashkeel
    
    text = strip_tashkeel(text)
    # text = qalsadi.lemmatizer.Lemmatizer(text)
    # remove repeated
    text = re.sub(r'(.)\1+', r'\1\1', text)
    text = text.replace('اا', 'ا').replace('وو', 'و').replace('يي', 'ي').replace('رر', 'ر')
    text = re.sub(r'((?<=\A)|(?<=\s))هه(?=\s|\Z)', r'ضحك', text)
    text = re.sub(r'ـ', '', text)
    # remove punctuations
    text = re.sub(r'[^\w\sؠ-ي١-٩]', ' ', text) # replace \w with \d to filter english letters
    text = re.sub(r'_', ' ', text)
    text = re.sub(r'/n', ' ', text)
    # unifying similar letters
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub('علے', 'على ', text)
    text = re.sub(r'ے', '', text)
    text = re.sub("[ؠىي]", "ي", text)
    # text = re.sub("يئ", "يء", text)
    text = re.sub("[ئؽؾؿ]", "ئ", text)# ء
    # text = re.sub("ؤ", "ء", text)
    # text = re.sub("ة", "ه", text)
    text = re.sub("[ؼػگ]", "ك", text)
    # separate negation signs
    text = re.sub(r'ما(\w+)ش', r'ما \1', text)
    text = re.sub(r'((?<=\A)|(?<=\s)|(?<=و))مش(?=\s)', r'ما', text)
    text = re.sub(r'((?<=\A)|(?<=\s)|(?<=و))ليش(?=\s)', r'ليه', text)
    text = re.sub(r'((?<=\A)|(?<=\s)|(?<=و))يا(?=\w)', r'يا ', text)
    text = re.sub('شاءالله', ' شاء الله', text)
    # remove slang begin marks
    text = re.sub(r'((?<=\A)|(?<=\s))ه(?=[ايت])', '', text)
#     text = re.sub('((?<=\s)|(?<=\A))(\w?ال)(?=\w{3,})', '', text)
#     text = re.sub('((?<=\s)|(?<=\A))(لل)(?=\w{3,})', '', text)
    # text = re.sub(r'((?<=\A)|(?<=\s))ب(?=[ايت])', '', text)
    # text = re.sub(r'((?<=\A)|(?<=\s))ا(?=[يت])', '', text)
    return " ".join(text.split())
```

```python
#load Model word2Vec Model
from gensim.models import KeyedVectors
word2vec = KeyedVectors.load_word2vec_format('w2v_model.bin', binary=True)
num_features = word2vec.vector_size
num_features   #vector length
```

```python
class WordVecVectorizer(object):
    def __init__(self, word2vec):
        self.word2vec = word2vec
        self.dim = word2vec.vector_size

    def fit(self, X, y):
        return self    

    def transform(self, X):
        return np.array([
            np.mean([self.word2vec[w] for w in texts.split() if w in self.word2vec]
                    or [np.zeros(self.dim)], axis=0)for texts in X])
```

```python
#representing each headline by the mean of word embeddings for the words used in the headlines.
wtv_vect = WordVecVectorizer(word2vec)
X_train_wtv = wtv_vect.transform(MedicalConsult.message)
X_train_wtv.shape
```

```python
#apply PCA
#Create a Covariance Matrix
covar_matrix = PCA(n_components = 300) #we have 300 features
```

```python
#Calculate Eigenvalues
covar_matrix.fit(X_train_wtv)
variance = covar_matrix.explained_variance_ratio_ #calculate variance ratios
var=np.cumsum(np.round(covar_matrix.explained_variance_ratio_, decimals=3)*100)
var #cumulative sum of variance explained with [n] features #take 180  90%
```

```python
plt.ylabel('% Variance Explained')
plt.xlabel('# of Features')
plt.title('PCA Analysis')
plt.ylim(30,100.5)
plt.style.context('seaborn-whitegrid')


plt.plot(var)
```

```python
#Using PCA reduction by 180 features
X_train_wtv_reduced = PCA(n_components=180)
X_train_wtv_ar_reduced=X_train_wtv_reduced.fit_transform(X_train_wtv)
```

```python
# k means determine k
distortions = []
K = range(1,15)
for k in K:
    kmeanModel = cluster.KMeans(n_clusters=k,random_state=42).fit(X_train_wtv_ar_reduced)
    distortions.append(sum(np.min(cdist(X_train_wtv_ar_reduced, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X_train_wtv_ar_reduced.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
km_araPCA = cluster.KMeans(n_clusters=3,random_state=42)
y_cluster_kmeans_PCA = km_araPCA.fit_predict(X_train_wtv_ar_reduced)
```

```python
#Calulation of Co-occurrence matrix

# to obtain the co-occureence matrix using top 6000 features of TFIDF vectorizer
cooccurrenceMatrix = np.zeros((len(top_6000_feat_names),len(top_6000_feat_names))) # co-occurance matrix
context_window = 2 # context window for co-occurance matrix
```

```python
#Calulating the co - occurance matrix
# Program to calculate the co-occurance matrix for each review.
for sent in tqdm(class0List): # obtaining each sentence of review
    words_sent = sent.split() # obtaining each word of a sent
    for index,word in enumerate(words_sent): # obtaining the index of each word from sentence
        if word in top_6000_feat_names: # checking whether the word is present in Top 6000
            for j in range(max(index - context_window,0),min(index + context_window , len(words_sent) -1) + 1):
                if words_sent[j] in top_6000_feat_names:
                    if words_sent[j] != word:
                        cooccurrenceMatrix[top_6000_feat_names.index(words_sent[j]), top_6000_feat_names.index(word)] += 1    
                    else:
                        pass
                else:
                    pass
        else:
            pass
```

```python
# Use Gensim's NMF to get the best num of topics via coherence score
texts = allwords
# Create a dictionary
# In gensim a dictionary is a mapping between words and their integer id
dictionary = Dictionary(texts)
#filter out tokens by their frequency to limit the number of features
dictionary.filter_extremes(  
    no_below=3,
    no_above=0.85,
    keep_n=5000
)
#ref:https://radimrehurek.com/gensim/corpora/dictionary.html#gensim.corpora.dictionary.Dictionary.filter_extremes
```

```python
# Create the bag-of-words format (list of (token_id, token_count))
corpus = [dictionary.doc2bow(text) for text in texts]
# Create a list of the topic numbers we want to try
topic_nums = list(np.arange(1, 10, 1))
```

**What is topic coherence?**
- Topic Coherence measures score a single topic by measuring the degree of semantic similarity between high scoring words in the topic.

**Coherence Measures**
**Let’s take quick look at different coherence measures, and how they are calculated:**
- **C_v**  It measures how often the topic words appear together in the corpus. 


- **C_p** is based on a sliding window, one-preceding segmentation of the top words and the confirmation measure of Fitelson’s coherence


- **C_uci** measure is based on a sliding window and the pointwise mutual information (PMI) of all word pairs of the given top words is represented as a bag of highly co-occurring context words that are weighted using either PMI 


- **C_umass** is based on document cooccurrence counts, a one-preceding segmentation and a logarithmic conditional probability as confirmation measure


- **C_npmi** is an enhanced version of the C_uci coherence Each topic word is represented as a bag of highly co-occurring context words that are weighted using either PMI or a normalised version of PMI (NPMI).


- **C_a** is baseed on a context window, a pairwise comparison of the top words and an indirect confirmation measure that uses normalized pointwise mutual information (NPMI) and the cosine similarity

[referance](https://towardsdatascience.com/evaluate-topic-model-in-python-latent-dirichlet-allocation-lda-7d57484bb5d0)

```python
coherence_scores = []

for num in topic_nums:
    nmf = Nmf(
        corpus=corpus,
        num_topics=num,
        id2word=dictionary,
        chunksize=2000,
        passes=5,
        kappa=.1,
        minimum_probability=0.01,
        w_max_iter=300,
        w_stop_condition=0.0001,
        h_max_iter=100,
        h_stop_condition=0.001,
        eval_every=10,
        normalize=True,
        random_state=42
    )
    
    # Run the coherence model to get the score
    cm = CoherenceModel(
        model=nmf,
        texts=texts,
        dictionary=dictionary,
        coherence='c_npmi'
    )
    
    coherence_scores.append(round(cm.get_coherence(), 5))

# Get the number of topics with the highest coherence score
scores = list(zip(topic_nums, coherence_scores))
best_num_topics = sorted(scores, key=itemgetter(1), reverse=True)[0][0]

# Plot the results
fig = plt.figure(figsize=(15, 7))

plt.plot(
    topic_nums,
    coherence_scores,
    linewidth=3,
    color='#4287f5'
)

plt.xlabel("Topic Num", fontsize=14)
plt.ylabel("Coherence Score", fontsize=14)
plt.title('Coherence Score by Topic Number - Best Number of Topics: {}'.format(best_num_topics), fontsize=18)
plt.xticks(np.arange(5, max(topic_nums) + 1, 5), fontsize=12)
plt.yticks(fontsize=12)

file_name = 'c_score'


plt.show()
```

```python
def top_words(topic, n_top_words):
    return topic.argsort()[:-n_top_words - 1:-1]
def topic_table(model, feature_names, n_top_words):
    topics = {}
    for topic_idx, topic in enumerate(model.components_):
        t = ("topic_%d:" % topic_idx)
        topics[t] = [feature_names[i] for i in top_words(topic, n_top_words)]
    return pd.DataFrame(topics)
```

```python

#plot word cloud function
import arabic_reshaper
from bidi.algorithm import get_display
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt


def plot_wordCloud_BiGram(mylistWords):
    mylistWords=list(map(lambda text: text.replace('ة', 'ه'), mylistWords))
    reshaped_list=[]
    for i in range(len(mylistWords)):
        reshaped_word = arabic_reshaper.reshape(mylistWords[i])
        reshaped_list.append(get_display(reshaped_word))
    word_could_dict=Counter(reshaped_list)
    word_cloud = WordCloud(font_path='arial', background_color="white", max_font_size=100, max_words=200,width=800,height=400)
    word_cloud = word_cloud.generate_from_frequencies(word_could_dict)
    plt.imshow(word_cloud, interpolation='bilinear')
    plt.axis("off")
    plt.rcParams["figure.figsize"] = (30,15)

    plt.show()
```
